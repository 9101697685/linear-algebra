```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
#panda data
data = pd.read_csv('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/Data File/2022-23-Intro-to-SSE-dataset-Assignment-2-LA.csv')
print(data.shape)
print(data.head())
```

```python
# collecting X and Y
X = data['Temperature (degree Celsius)'].values
Y = data['Output Raspberry Pie'].values

# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
n = len(X)

# Using the formula to calculate 'a and 'b'
number = 0
denom = 0
for i in range(n):
    number += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
a = number / denom
b = mean_y - (a * mean_x)

# Printing coefficients
print ("coefiicients")
print (a, b)

# Plotting Values and Regression Line
max_x = np.max(X) + 5
min_x = np.min(X) - 5

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = b + a * x


# Ploting Line
plt.plot(x, y, color='#FF8000', label='y = ax + b')


# Ploting Data Points
plt.scatter(X, Y, color='#1874CD', label='data')


plt.xlabel('Temperature (degree Celsius)')
plt.ylabel('Output Raspberry Pie')
plt.legend(framealpha=1, shadow=True)
plt.grid(alpha=0.25)
plt.show()
```
