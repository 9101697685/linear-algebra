```python
from matplotlib.image import imread # image read
import matplotlib.pyplot as plt
import numpy as np
import os

A= imread ('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/Data File/car photo.jpg') # upload photo
X= np.mean(A, -1)
print(X.shape)

orignal_photo_size = os.path.getsize('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/Data File/car photo.jpg')
print(f"original photo size is: {orignal_photo_size//1024} KB")

plt.rcParams['figure.figsize'] = (8, 4)
img= plt.imshow(X)
img.set_cmap('gray')
plt.show()
```

```python
U, S, VT= np.linalg.svd(X, full_matrices=False) # full_matrices=False means economy svd
S = np.diag(S)

j = 0
for r in (1, 5, 10, 20, 60, 100, 200):
    Xapprox = U [:,:r] @ S[0:r, :r] @ VT[:r, :]
   
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.savefig('r = '+str(r)+'.jpg') 
    plt.figure(j+1)
    
    plt.axis('off')
    plt.title('rank = ' + str(r))
    plt.show()

   
    size = os.path.getsize('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/r = '+str(r)+'.jpg')
    print('r = '+str(r)+'.jpg size = ',size//1024 ,'KB')
    print('Compression ratio = ','%.2f' % (orignal_photo_size / size))
    
# U, S, VT= np.linalg.svd(X, full_matrices=False) # full_matrices=False means economy svd
# S = np.diag(S)

# j = 0
# for r in (0, 1, 5, 10, 20, 60, 100): #construct approximate image 
#     Xapprox = U [:,:r] @ S[0:r, :r] @ VT[:r, :]
    
    
    
#     plt.savefig('saved.jpeg') 
# #     from PIL import Image
# #     img = img.save('savedimage.jpg')
    
#     plt.figure(j+1)
    
    
#     img = plt.imshow(256-Xapprox)
#     img.set_cmap('gray')
#     plt.axis('off')
#     plt.title('rank= ' + str(r)+"      "+ 'ratio= '+"      "+ "size= ")
#     plt.show()
```

```python
U, S, VT= np.linalg.svd(X, full_matrices=False) # full_matrices=False means economy svd
S = np.diag(S)

j = 0
for r in (1, 5, 10, 20, 60, 100, 200):
    Xapprox = U [:,r:] @ S[r:, r:] @ VT[r:, :]
   
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.savefig('r = '+str(r)+'error_image.jpg') 
    plt.figure(j+1)
    
    plt.axis('off')
    plt.title('rank = ' + str(r))
    plt.show()

   
    size_image_error = os.path.getsize('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/r = '+str(r)+'error_image.jpg')
    print('r = '+str(r)+'.jpg size = ',size_image_error//1024 ,'KB')
    print('Compression ratio = ','%.2f' % (size_image_error / size*100))
```
