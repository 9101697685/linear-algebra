```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
#panda data
data = pd.read_csv('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/Data File/2022-23-Intro-to-SSE-dataset-Assignment-2-LA.csv')
print(data.shape)
print(data.head())
```

### a. Linear model ( y = ax + b )

```python
# collecting X and Y
X = data['Temperature (degree Celsius)'].values
Y = data['Output Raspberry Pie'].values

# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
n = len(X)

# Using the formula to calculate 'a and 'b'
number = 0
denom = 0
for i in range(n):
    number += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
a = number / denom
b = mean_y - (a * mean_x)

# Printing coefficients
print ("coefiicients")
print (a, b)

# Plotting Values and Regression Line
max_x = np.max(X) + 5
min_x = np.min(X) - 5

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = b + a * x


# Ploting Line
plt.plot(x, y, color='#FF8000', label='y = ax + b')


# Ploting Data Points
plt.scatter(X, Y, color='#1874CD', label='data')


plt.xlabel('Temperature (degree Celsius)')
plt.ylabel('Output Raspberry Pie')
plt.legend(framealpha=1, shadow=True)
plt.grid(alpha=0.25)
plt.show()
```

### b. Polynomial of the second order(y = ax^2 + bx + c)

```python
from scipy.linalg import lstsq
import matplotlib.pyplot as plt
M = X[:, np.newaxis]**[0, 2]
p, res, rnk, s = lstsq(M, Y)
```

```python
plt.plot(X, Y, 'o', label='data')
xx = np.linspace(-45, 45, 45)
yy = p[1]*xx**2 + p[1]*xx + p[0]
plt.plot(xx, yy, label='y = ax^2 + bx + c')
plt.xlabel('Temperature (degree Celsius)')
plt.ylabel('Output Raspberry Pie')
plt.legend(framealpha=1, shadow=True)
plt.grid(alpha=0.25)
plt.show()
```

### c. Polynomial of the third order (y = ax^3 + bx^2 + cx + d)

```python
from scipy.linalg import lstsq
import matplotlib.pyplot as plt
M = X[:, np.newaxis]**[0, 3]
p, res, rnk, s = lstsq(M, Y)
```

```python
plt.plot(X, Y, 'o', label='data')
xx = np.linspace(-45, 45, 45)
yy = p[1]*xx**3 + p[1]*xx**2 + p[1]*xx + p[0]
plt.plot(xx, yy, label='y = ax^3 + bx^2 + cx + d')
plt.xlabel('Temperature (degree Celsius)')
plt.ylabel('Output Raspberry Pie')
plt.legend(framealpha=1, shadow=True)
plt.grid(alpha=0.25)
plt.show()
```

```python
from matplotlib.image import imread # image read
import matplotlib.pyplot as plt
import numpy as np
import os

A= imread ('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/Data File/car photo.jpg') # upload photo
X= np.mean(A, -1)
print(X.shape)

orignal_photo_size = os.path.getsize('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/Data File/car photo.jpg')
print(f"original photo size is: {orignal_photo_size//1024} KB")

plt.rcParams['figure.figsize'] = (8, 4)
img= plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT= np.linalg.svd(X, full_matrices=False) # full_matrices=False means economy svd
S = np.diag(S)

j = 0
for r in (1, 5, 10, 20, 60, 100, 200):
    Xapprox = U [:,:r] @ S[0:r, :r] @ VT[:r, :]
   
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.savefig('r = '+str(r)+'.jpg') 
    plt.figure(j+1)
    
    plt.axis('off')
    plt.title('rank = ' + str(r))
    plt.show()

   
    size = os.path.getsize('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/r = '+str(r)+'.jpg')
    print('r = '+str(r)+'.jpg size = ',size//1024 ,'KB')
    print('Compression ratio = ','%.2f' % (orignal_photo_size / size))
```

```python
# Error Image

U, S, VT= np.linalg.svd(X, full_matrices=False) # full_matrices=False means economy svd
S = np.diag(S)

j = 0
for r in (1, 5, 10, 20, 60, 100, 200):
    Xapprox = U [:,r:] @ S[r:, r:] @ VT[r:, :]
   
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.savefig('r = '+str(r)+'error_image.jpg') 
    plt.figure(j+1)
    
    plt.axis('off')
    plt.title('rank = ' + str(r))
    plt.show()

   
    size_image_error = os.path.getsize('/Users/peter/Documents/Smart System/Semester 1/Project/Linear Algebra Project/r = '+str(r)+'error_image.jpg')
    print('r = '+str(r)+'.jpg size = ',size_image_error//1024 ,'KB')
    print('Error = ','%.2f' % (size_image_error / size*100))
```
