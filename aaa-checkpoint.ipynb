```python
from matplotlib.image import imread # image read
import matplotlib.pyplot as plt
import numpy as np
import os

A= imread ('/Users/peter/Downloads/Linear Algebra Project/car photo.jpg') # upload photo
X= np.mean(A, -1)
img= plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT= np.linalg.svd(X, full_matrices=False) # full_matrices=False means economy svd
S = np.diag(S)

j = 0
for r in (0, 2, 5, 10, 20, 60, 100): #construct approximate image 
    Xapprox = U [:,:r] @ S[0:r, :r] @ VT[:r, :]
    plt.figure(j+1)
    img = plt.imshow(256-Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

I will use the diagram below(Sigular Values: Cumulative Sum) to answer the rest of the question.

```python
plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Sigular Values: Cumulative Sum')
plt.show()
```
