```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
#panda data
data = pd.read_csv('/Users/peter/Downloads/Linear Algebra Project/2022-23-Intro-to-SSE-dataset-Assignment-2-LA.csv')
print(data.shape)
print(data.head())
```

## a. Linear model ( y = ax + b )

```python
# collecting X and Y
X = data['Temperature (degree Celsius)'].values
Y = data['Output Raspberry Pie'].values

# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Total number of values
n = len(X)

# Using the formula to calculate 'm and 'c'
number = 0
denom = 0
for i in range(n):
    number += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
m = number / denom
c = mean_y - (m * mean_x)

# Printing coefficients
print ("coefiicients")
print (m, c)

# Plotting Values and Regression Line
max_x = np.max(X) + 5
min_x = np.min(X) - 5

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = c + m * x


# Ploting Line
plt.plot(x, y, color='#58b970', label='Regression Line')


# Ploting Satter Points
plt.scatter(X, Y, color='#ef5423', label='Scatter plot')


plt.xlabel('Temperature (degree Celsius)')
plt.ylabel('Output Raspberry Pie')
plt.legend()
plt.show()
```
